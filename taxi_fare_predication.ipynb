```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/taxi_fare/train.csv")
df = df[df["distance_traveled"]<=150]
df
```

```python
#plotdata = df.iloc[rowstart:rowend,columnstart:columnend]
x = df.drop(columns="total_fare")
x
```

```python
y = df["total_fare"]
y
```

```python
for col in df.columns:
    plt.scatter(x=df[col],y=df["total_fare"])
    plt.xlabel(col)
    plt.ylabel("Total Fare")
    plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train,x_test, y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
lin_reg_model = LinearRegression()
lin_reg_model.fit(x_train,y_train)
```

```python
test_predict = lin_reg_model.predict(x_test)
test_predict
```

```python
x_test.shape[0]
x_index = np.arange(0,x_test.shape[0])
```

```python
plt.scatter(x_index,y_test,alpha=1.0, c="green")
plt.scatter(x_index,test_predict,alpha=0.3, c= "red")
plt.show()
```
